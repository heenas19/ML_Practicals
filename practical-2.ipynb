# To convert a list of numeric values into a one-dimensional NumPy array

```python
import numpy as np

l = [1,2,3,4,5]

a= np.array(l)

print(a)
```

# To create a 3x3 matrix with values ranging from 2 to 10

```python
import numpy as np

a = np.arange(2,11).reshape(3,3)

print(a)
```

# To append values at the end of an array

```python
import numpy as np

a1 = np.random.randint(0,10,5)

a2 = np.random.randint(0,10,5)

a = np.append(a1, a2)

print('Original Array : ', a1)

print('\nArray after appending : ', a)
```

# To create another shape from an array without changing its data(3*2 to 2*3) 

```python
import numpy as np

a = np.random.randint(1,10,(3,2))

print("Original Array : \n",a)

b = a.reshape(2,3)

print("\nNew Shape : \n",b)
```
