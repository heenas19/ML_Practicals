# Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/BAJAJ-AUTO.csv')

start_date = '2008-05-26'
end_date = '2009-05-26'

df['Date'] = pd.to_datetime(df['Date'])

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize = (11,6))
plt.plot(filtered_df['Date'],filtered_df['Open'])
plt.plot(filtered_df['Date'],filtered_df['Close'])

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
plt.show()
```
