# To split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order

```python
import numpy as np

a = np.random.randint(0,10,14)

a1, a2, a3 = np.split(a,[2,6])

print(a1, a2, a3)
```

# To stack arrays horizontally (column wise)

```python
import numpy as np

a1 = np.random.randint(0,10, 5)

a2 = np.random.randint(0,10, 5)

a = np.hstack([a1,a2])

print(a)
```
