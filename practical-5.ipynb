# To find the maximum and minimum value of a given flattened array

```python
import numpy as np

a = np.random.randint(10,20,(3,3))

f = a.flatten()

print("\nFlattened Array : ",f)

print("\nMaximum Value is : ",np.max(f))

print("\nMinimum Value is : ",np.min(f))
```

# To compute the mean, standard deviation, and variance of a given array along the second axis

```python
import numpy as np

a = np.random.randint(5,15,(2,3))

m = np.mean(a,axis=1)

print("\nMean : ",m)

s = np.std(a,axis=1)

print("\nStandard Deviation : ",s)

v = np.var(a,axis=1)

print("\nVariance",v)
```
